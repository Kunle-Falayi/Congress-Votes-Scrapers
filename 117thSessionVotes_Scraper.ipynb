```python
base_url = "https://clerk.house.gov/Votes/2021{}"
print("Scraping URL:", base_url)
time.sleep(1.5)
master_data = []
for i in range(1, 450):
    url = base_url.format(i)
    page = requests.get(url)
    soup_doc = bs(page.content, "html.parser")

    tables = soup_doc.find_all('tbody', {'id': 'member-votes'})

    # Find the bill number element and extract its text
    bill_number_element = soup_doc.find('span', {'class': 'legisNum'})
    bill_number = bill_number_element.text if bill_number_element else ''
    #Find the bill details and extract text
    bill_details_element = soup_doc.find('p', {'class': 'roll-call-description'})
    bill_details = bill_details_element.text if bill_details_element else ''

    for table in tables:
        rows = table.find_all('tr')
        list_headers = ['Member', 'Rep', 'Party', 'State', 'State_Abbr', 'Vote', 'Bill Number', 'Bill Details']
        for row in rows:
            vote_list = row.find_all('td')
            vote_data = [col.text for col in vote_list]
            vote_data.append(bill_number)
            vote_data.append(bill_details)
            vote_dict = dict(zip(list_headers, vote_data))
            master_data.append(vote_dict)

print(master_data)
```

```python
pd.option_context('display.max_colwidth', None)
df = pd.DataFrame(master_data)
df.to_csv('108th_1stndSession.csv')
df.head()
```
